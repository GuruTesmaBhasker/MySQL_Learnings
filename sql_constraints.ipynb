```python
%%sql
/* 
   PRIMARY KEY Explanation:

   - A Primary Key is a column (or a group of columns) that uniquely 
     identifies each row in a table.

   - It must always be UNIQUE:
       No two rows can have the same primary key value.

   - It must always be NOT NULL:
       Every row must contain a value for the primary key.

   - A table can have only ONE Primary Key constraint, 
     but that key can include multiple columns (composite key).

   - The Primary Key helps maintain data integrity and allows 
     the database to quickly locate specific records.

   - It is automatically indexed by the database for faster search.
*/
```

```python
%%sql
/* 
   COMPOSITE PRIMARY KEY (ride_id, rider_id) – Explanation:

   - In this rides table, neither ride_id nor rider_id alone can uniquely
     identify a row because both columns contain duplicate values.
         * Example: ride_id = 1 appears multiple times.
         * Example: rider_id = 202 appears multiple times.

   - However, the COMBINATION of (ride_id, rider_id) is always unique.
     This makes them a perfect pair for a composite primary key.

   - This composite key ensures:
       1. No duplicate ride–rider pair can exist.
       2. The correct mapping of which rider booked which ride.
       3. Strong data integrity for ride booking relationships.

   - In real-world ride-sharing databases:
       * A ride can have multiple riders (carpool).
       * A rider can join different rides.
       * So a single-column primary key is not enough.

   - Hence, using PRIMARY KEY (ride_id, rider_id) fits naturally 
     with the structure and meaning of this data.


    GENERAL EXPLNATION 
     
   COMPOSITE PRIMARY KEY EXPLANATION:

   - A composite primary key is created using two or more columns together.
   - These columns, when combined, uniquely identify each row in the table.
   - Individually, each column may contain duplicate values,
     but the combination of values across the key columns must be unique.

   - All columns in a composite primary key are automatically NOT NULL.
   - This ensures data integrity and prevents duplicate or incomplete records.

   - Composite keys are useful when:
       * No single column can uniquely identify a row.
       * The uniqueness depends on a combination of attributes
         (e.g., ride_id + rider_id, order_id + product_id).

   - The database automatically creates an index on the composite key,
     improving search and lookup performance based on these columns.

*/
```

```python
%%sql
/* 
   UNIQUE Constraint – Explanation:

   - A UNIQUE constraint ensures that the values in a column 
     (or a group of columns) do not repeat.
     Every value must be different from every other value.

   - Unlike a PRIMARY KEY:
        * UNIQUE allows only one NULL value (in most SQL databases)
        * A table can have multiple UNIQUE constraints
        * PRIMARY KEY = UNIQUE + NOT NULL (stronger rule)

   - UNIQUE is used when:
        * A column should not repeat, but it doesn’t qualify as a primary key.
        * Example: email, phone_number, username, license_number, etc.

   - A UNIQUE constraint helps prevent accidental duplicate entries 
     and maintains clean, reliable data.
*/
```

```python
%%sql
CREATE TABLE rides1 (
    ride_id INT PRIMARY KEY,
    driver_id INT NOT NULL,
    rider_id INT,
    pickup_location VARCHAR(100) NOT NULL,
    dropoff_location VARCHAR(100) NOT NULL,
    ride_date DATETIME NOT NULL,
    fare DECIMAL(10, 2) CHECK (fare > 0 ) #check constraints 
    );
```

```python
%%sql
INSERT INTO rides1 (ride_id, driver_id, rider_id, pickup_location, dropoff_location, ride_date, fare)
VALUES
(3, 101, 201, 'Chennai', 'Coimbatore', '2024-12-29 08:00:00', 912.8);
# it cannot be - value, because we gave the conditon above. it works like a where function 
```

```python
%%sql
select * from rides1;
```

```python
%%sql
/* 
   CHECK Constraint:
   - Ensures that values entered in a column satisfy a specific condition.
   - Helps enforce valid and logical data (e.g., fare > 0, age >= 18).
   - Prevents invalid or out-of-range values from being inserted.
*/
```

```python
%%sql
-- Creating the drivers table

drop table rides2;
drop table drivers;
CREATE TABLE drivers (
    driver_id INT PRIMARY KEY,
    driver_name VARCHAR(100),
    license_number VARCHAR(50) UNIQUE
);

-- Creating the rides table with driver_id as a foreign key
CREATE TABLE rides2 (
    ride_id INT PRIMARY KEY,
    driver_id INT,
    pickup_location VARCHAR(100),
    dropoff_location VARCHAR(100),
    ride_date DATETIME,
    fare DECIMAL(10, 2),
    FOREIGN KEY (driver_id) REFERENCES drivers(driver_id)
);
```
