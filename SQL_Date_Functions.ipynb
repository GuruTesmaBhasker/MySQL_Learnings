```python
%%sql
Drop table if exists events;
```

```python
%%sql
CREATE TABLE events (
    event_id INT AUTO_INCREMENT PRIMARY KEY,
    event_name VARCHAR(100),
    event_date DATETIME,
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);
```

```python
%%sql
INSERT INTO events (event_name, event_date)
VALUES
('Team Meeting',  '2025-01-15 10:30:00'),
('Conference',    '2025-02-10 09:00:00'),
('Online Webinar','2025-03-05 15:45:00'),
('Workshop',      '2025-03-20 13:30:00');
```

```python
%%sql
select * from events
```

```python
%%sql
# Date and time 
# Extracting the year from the event column 
select event_id, event_name, YEAR(event_date) as Year_event, Date(event_date) as date_event, day(event_date) as day_event from events 
# the YEAR, date, dayfunction does this 
```

```python
%%sql
select event_id, event_name,event_date, event_date + INTERVAL 30 day as Added_30_days from events
# Adds the 30 or n days you want to add in the actual date 
```

```python
%%sql
select event_id, event_name,event_date, event_date - INTERVAL 30 day as Sub_30_days from events 
# Minus 30 days from the actual date 
```

```python
%%sql
select event_id, event_name, Date_format(event_date,'%W, %M %e %Y') as formatted_dates from events;
```

```python
%%sql
select event_id, event_name, Date_format(event_date,'%h %i %p') as formatted_dates from events;
```
