```python
%%sql
Drop table if exists accounts;
CREATE TABLE Accounts (
  AccountID INT PRIMARY KEY,
  AccountHolder VARCHAR(100),
  Balance DECIMAL(10,2)
);

-- Insert a new record into Accounts
INSERT INTO Accounts (AccountID, AccountHolder, Balance)
VALUES (1, 'Alice', 5000.00);
```

```python
%%sql
select * from accounts
```

```python
%%sql
START TRANSACTION;
```

```python
%%sql
-- Perform one or more SQL statements
UPDATE Accounts SET Balance = Balance - 1000 WHERE AccountID = 1;
```

```python
%%sql
select * from accounts
```

```python
%%sql
rollback;
```

```python
%%sql
# The Balance will be 5000 only beacuse we did roll back
# Here in the notebooks it is not working but in command line works
select * from accounts
```

```python
%%sql
# After roll back we can update the things 
# After updating we can commit it and we cannot able to roll back after we commit
UPDATE Accounts SET Balance = Balance - 3000 WHERE AccountID = 1;
```
