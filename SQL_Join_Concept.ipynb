```python
%%sql
-- we need restaurant name and order date to reporting team
-- we are going to use INNER JOIN for this report
select
r.name as restaurant_name,
o.order_date
from Restaurants r
join orders o on r.id = o.restaurant_id
```

```python
%%sql
# LEFT JOIN
select
r.name as restaurant_name,
o.order_date
from Restaurants r
left join orders o on r.id = o.restaurant_id
```

```python
%%sql
# RIGHT JOIN
select 
r.name as restaurant_name,
o.order_date
from Restaurants r
right join orders o on r.id = o.restaurant_id
```

```python
%%sql
# SELF JOIN to check orders on the same date, 
SELECT 
    a.order_id AS order_one,
    b.order_id AS order_two,
    a.order_date
FROM Orders a
JOIN Orders b
    ON a.order_date = b.order_date
    AND a.order_id <> b.order_id;
```
