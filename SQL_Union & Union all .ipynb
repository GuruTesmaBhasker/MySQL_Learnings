```python
%%sql
-- Drop tables if they already exist (prevents errors when recreating them)
DROP TABLE IF EXISTS employees;
DROP TABLE IF EXISTS contractors;

-- Creating the employees table
CREATE TABLE employees (
    employee_id INT AUTO_INCREMENT PRIMARY KEY,  -- Unique ID, auto-generated for each employee
    first_name VARCHAR(50),                      -- Employee's first name
    last_name VARCHAR(50),                       -- Employee's last name
    position VARCHAR(50),                        -- Job title or role
    salary DECIMAL(10,2)                         -- Salary stored as a decimal number (2 decimal places)
);

-- Creating the contractors table
CREATE TABLE contractors (
    contractor_id INT AUTO_INCREMENT PRIMARY KEY, -- Unique ID, auto-generated for each contractor
    first_name VARCHAR(50),                       -- Contractor's first name
    last_name VARCHAR(50),                        -- Contractor's last name
    position VARCHAR(50),                         -- Job role or designation
    hourly_rate DECIMAL(10,2)                     -- Payment rate per hour for contractors
);
```

```python
%%sql
-- UNION combines results from both tables and removes duplicates

SELECT first_name, last_name, position
FROM employees

UNION

SELECT first_name, last_name, position
FROM contractors;
```

```python
%%sql
-- UNION ALL keeps duplicates

SELECT first_name, last_name, position
FROM employees

UNION ALL

SELECT first_name, last_name, position
FROM contractors;
```

```python
%%sql
-- UNION      → combines results & removes duplicates
-- UNION ALL  → combines results & keeps duplicates
```
