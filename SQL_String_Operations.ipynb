```python
%%sql
CREATE TABLE Customersdetails (
    id INT PRIMARY KEY,
    customer_name VARCHAR(50),
    email VARCHAR(100),
    city VARCHAR(50),
    phone_number VARCHAR(20)
);
```

```python
%%sql 
select * from customersdetails
```

```python
%%sql # Length function
select length (customer_name) as name_length,
char_length(customer_name) as Char_len
from customersdetails
```

```python
%%sql # Upper and Lower cas3
SELECT Id,
       customer_name,
       UPPER(customer_name) AS name_upper,
       LOWER(customer_name) AS name_lower
FROM Customersdetails;
```

```python
%%sql
-- Join name + city into one string
SELECT 
    customer_name,
    city,
    CONCAT(customer_name, ' from ', city) AS full_info
FROM Customersdetails;
```

```python
%%sql
-- Combine values with a separator, skipping NULL phone_number
SELECT 
    CONCAT_WS(' | ', customer_name, email, phone_number) AS details
FROM Customersdetails;
```

```python
%%sql
-- Extract parts of strings
SELECT
    customer_name,
    SUBSTRING(customer_name, 1, 4) AS first_4_letters,
    LEFT(city, 3) AS city_first3,
    RIGHT(phone_number, 2) AS last_2_digits
FROM Customersdetails;
```

```python
%%sql
-- Replace email domain
SELECT
    email,
    REPLACE(email, '@example.com', '@gmail.com') AS updated_email
FROM Customersdetails;
```

```python
%%sql
-- Remove spaces from city names
SELECT
    city,
    TRIM(city) AS trim_city,# IT removes empty spaces from the column, checks both side
    LTRIM(city) AS left_trim,# it also does the same but check space only in left
    RTRIM(city) AS right_trim# it also does the same but check space only in right
FROM Customersdetails;
```

```python
%%sql
-- Find position of '@' in emails
SELECT
    email,
    LOCATE('@', email) AS at_position
FROM Customersdetails;
```

```python
%%sql
-- Get customers whose name starts with 'A'
SELECT * # LIke (pattern search)
FROM Customersdetails
WHERE customer_name LIKE 'A%' # the percentage states anything comes after A is not matter but i need the word from A 
/* 
    %   = matches ANY number of characters
    'A%'   = starts with A
    '%A'   = ends with A
    '%A%'  = contains A anywhere
    '_A%'  = second letter is A ( _ = matches exactly one character )
*/
```
